# Simple Linear Regression

## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Generating Random Data

```python
num_points = 50

x = np.random.randint(0, 100, num_points)
y = 2 * x + np.random.normal(0, 10, num_points)
```

## Creating Data Frame and Saving to CSV

```python
data = pd.DataFrame({'X': x, 'Y': y})

data.to_csv('data.csv', index=False)
print(data.head())
```

## Visualizing the Data

```python
plt.scatter(data['X'], data['Y'], color="crimson")
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (Y)")
plt.title("Scatter Plot of Data")
plt.show()
```

## Preparing for Data Modeling

```python
X = data[['X']] #Independent Variable
Y = data[['Y']] # Dependent Variable
```

## Fitting the Linear Regression Model

```python
model = LinearRegression()
model.fit(X, Y)
```

## Getting Model Coefficients

```python
slope = model.coef_[0]
intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)
```

## Visualizing the Linear Regression Line

```python
plt.scatter(X, Y)
plt.plot(X, model.predict(X), color='skyblue')
plt.xlabel("Independent Variable (X)")
plt.ylabel("Dependent Variable (Y)")
plt.title("Linear Regression")
plt.show()
```

## Making Predictions

```python
new_X = [[np.random.randint(1, 200)]]
predicted_Y = model.predict(new_X)
print('Value of X:', new_X)
print('Predicted Value of Y:', predicted_Y)
```
