# Time Series Analysis using ARIMA Model

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters

register_matplotlib_converters()
```

## 2. Load Dataset

```python
data = pd.read_csv("sales-of-shampoo-over-a-three-year.csv")
data.describe()
```

## Convert Month to Datetime

```python
data['Month'] = pd.date_range(start = '2000-01-01', periods = len(data), freq = 'M')
data.set_index('Month', inplace = True)
```

## Plot the Time Series

```python
plt.figure(figsize = (10, 6))
plt.plot(data.index, data['Sales of shampoo over a three year period'], marker = 'o', linestyle='--', color = 'darkcyan')
plt.title("Sales of Data Shampoo Over Time")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation = 45)
plt.grid(True)
plt.show()
```

## Decomposition

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(data['Sales of shampoo over a three year period'], model = 'additive')
fig = decomposition.plot()
plt.show()
```

## Forecasting with ARIMA

```python
from statsmodels.tsa.stattools import adfuller
adf_test = adfuller(data['Sales of shampoo over a three year period'])
print('ADF Statistics', adf_test[0])
print('P-value =', adf_test[1])
```

## Fit an ARIMA Model

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(data['Sales of shampoo over a three year period'], order = (1, 1, 1))
results = model.fit()

print(results.summary())
```

## Forecasting the Future Values

```python
forecast = results.forecast(steps = 8)
plt.figure(figsize = (10, 6))
plt.plot(data.index, data['Sales of shampoo over a three year period'], label = 'Historical', marker = 'o', linestyle='--', color = 'darkcyan')
plt.plot(forecast.index, forecast, label = 'Forecast', marker = 'o', linestyle='--', color = 'darkturquoise')
plt.grid(True)
plt.legend()
plt.show()
```
